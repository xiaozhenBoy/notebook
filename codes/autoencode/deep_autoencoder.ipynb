```python
"""
Deep autoencoder
"""
from keras.layers import Input, Dense
from keras.models import Model

from keras.datasets import mnist
import numpy as np

import matplotlib.pyplot as plt

input_size = 784
# the size of encoded representation
encoding_dim = 32

input_img = Input(shape=(input_size, ))
# encode layer
encode_layer1 = Dense(128, activation='relu')(input_img)
encode_layer2 = Dense(64, activation='relu')(encode_layer1)
encode_layer3 = Dense(encoding_dim, activation='relu')(encode_layer2)

# decode layer
decode_layer1 = Dense(64, activation='relu')(encode_layer3)
decode_layer2 = Dense(128, activation='relu')(decode_layer1)
decode_layer3 = Dense(784, activation='sigmoid')(decode_layer2)

# Model
autoencode = Model(input=input_img, output=decode_layer3)

####*************seperated part***************####
# encode model
encode_model = Model(input=input_img, output=encode_layer3)

# decode model
# decode_input = Input(shape=(encoding_dim, )) # decoder = autoencode.layers[-1]
decode_model = Model(input=input_img, output=decode_layer3)

# compile model
autoencode.compile(optimizer='adadelta', loss='binary_crossentropy')

print('Finish building model!')

# load data
print('Loading data ...')
(x_train, _), (x_test, _) = mnist.load_data()
x_train = x_train.astype('float32')/255.
x_test = x_test.astype('float32')/255.
x_train = x_train.reshape(len(x_train), np.prod(x_train.shape[1:])) # reshape from 28*28 to 1* 784
x_test = x_test.reshape(len(x_test), np.prod(x_test.shape[1:]))

print 'Training ...'

autoencode.fit(x_train, x_train, nb_epoch=40, batch_size=256, shuffle=True, validation_data=(x_test, x_test))

# encode and decode
encode_imgs = encode_model.predict(x_test)
decode_imgs = decode_model.predict(x_test)

n = 10
plt.figure(figsize=(20, 4))
for i in range(n):
    ax = plt.subplot(2, n, i)
    plt.imshow(x_test[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    
    ax = plt.subplot(2, n, i+n)
    plt.imshow(decode_imgs[i].reshape(28, 28))
    plt.gray()
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()
```
